## Limit of a Sequence

A sequence $X = \langle x_n \rangle$ in $\mathbb{R}$ is said to ***converge*** to $x \in \mathbb{R}$, or $x$ is said to be a ***limit of*** $\langle x_n \rangle$, if for every $\epsilon > 0$, there exists a natural number $K(\epsilon)$ such that for all $n \geq K(\epsilon)$, the term $x_n$ satisfy $|x_n - x| < \epsilon$.

- The notation $K(\epsilon)$ is used to emphasize that the choice of $K$ depends to the value of $\epsilon$ (We can use $k = K(\epsilon)$)
- When a sequence has a limit $x$, we will use the notation $\lim X = x$ or $\lim \langle x_n \rangle = x$.

## Theorem 1 *Uiqueness of Limits*

A sequence in $\mathbb{R}$ can have at most one limit.

**Proof**

Let $\langle x_n \rangle$ be a sequence in $\mathbb{R}$. Suppose that $x_1$ and $x_2$ are both limits of $\langle x_n \rangle$. For each $\epsilon > 0$ there exists $k_1$ such that $|x_n - x_1| < \frac{\epsilon}{2}$ for all $n \geq k_1$, and there exists also $k_2$ such that $|x_n - x_2| < \frac{\epsilon}{2}$ for all $n \geq k_2$. Let $k = \max (k_1, k_2)$. Then for $n \geq k$ we apply the Triangle Inequality, that is,

$$
\begin{align}
|x_1 - x_2| & = |x_1 - x_n + x_n - x_2| \\
& \leq |x_1 - x_n| + |x_n - x_2| \\
& \leq \frac{\epsilon}{2} + \frac{\epsilon}{2} = \epsilon
\end{align}
$$

Since $\epsilon > 0$ is arbitrary positive number, it follows that $x_1 - x_2 = 0$. Thus, $x_1 = x_2$. Hence, $\lim \langle x_n \rangle = x_1 = x_2$.

## Theorem 2

Let $X = \langle x_n \rangle$ be a sequence of real numbers, and let $x \in \mathbb{R}$. The following statements are equivalent.

- $X$ converges to $x$.
- For every $\epsilon > 0$, there exists a natural number $k$ such that for all $n \geq k$, the terms $x_n$ satisfy $|x_n - x| < \epsilon$.
- For every $\epsilon > 0$, there exists a natural number $k$ such that for all $n \geq k$, the terms $x_n$ satisfy $x - \epsilon < x_n < x + \epsilon$.
- For every $\epsilon$-neighborhood $V_{\epsilon}(x)$ of $x$, there exists a naturla number $k$ such that for all $n \geq k$, the terms $x_n$ belongs to $V_{\epsilon}(x)$.

**Proof**

Let $\langle x_n \rangle$ be a sequence in $\mathbb{R}$, and let $x \in \mathbb{R}$.

- Directly follows from the definition of convergent.
- Let $\epsilon > 0$ such that for some $k \in \mathbb{N}$, $|x_n - x| < \epsilon$ for all $n \geq k$ $(n \in \mathbb{N})$ Now,

$$
\begin{align}
|x_n - x| < \epsilon & \implies -\epsilon < x_n - 1 < \epsilon \\
& \implies x - \epsilon < x_n < x + \epsilon
\end{align}
$$

- Let $\epsilon > 0$ such that for some $k \in \mathbb{N}$, $x - \epsilon < x_n < x + \epsilon$ for all $n \geq k$ $(n \in \mathbb{N}$. Now,

$$
\begin{align}
x - \epsilon < x_n < x + \epsilon & \implies -\epsilon < x_n - x < \epsilon \\
& \implies |x_n - x| < \epsilon
\end{align}
$$

Hence, $x_n \in V_{\epsilon}(x)$ for all $n$.

- Let $\epsilon > 0$. Let $x_n \in V_{\epsilon}(x)$. Then $x - \epsilon < x_n < x + \epsilon$. Thus, $-\epsilon < x_n - x < \epsilon$. Hence, $|x_n - x| < \epsilon$. Therefore, $\langle x_n \rangle$ converges to $x$.

## Tail of Sequences

If $X = \langle x_1, x_2, \dots, x_n, \dots \rangle$ is a sequence of real numbers and if $m$ is a given natural number, then the $m$-**tail** of $X$ is the sequence

$$
X_m := \langle x_{m + n} : n \in \mathbb{N} \rangle = \langle x_{m + 1}, x_{m + 2}, \dots \rangle
$$

## Theorem 3

Let $X = \langle x_n : n \in \mathbb{N} \rangle$ be a sequence of real numbers and let $m \in \mathbb{N}$. Then the $m$-tail $X_m = \langle x_{m + n} : n \in \mathbb{N}$ of $X$ converges if and only if $X$ converges. In this case, $\lim X_m = \lim X$.

## Limit Theorems

A sequence $X = \langle x_n \rangle$ of real numbers is said to be bounded if there exists a real number $M > 0$ such that $|x_n| \leq M$ for all $n \in N$.

## Theorem 4

A convergent sequence of real numbers is bounded.

**Proof**

Suppose that $\lim \langle x_n \rangle = x$ and let $\epsilon = 1$. Then there exists $k \in N$ such that $|x_n - x| < 1$ for all $n \geq k$. Now,

$$
\begin{align}
|x_n| & = |x_n - x + x| \\
& \leq |x_n - x| + |x| \\
& < 1 + |x|
\end{align}
$$

Thus, $|x_n| < 1 + |x|$ for all $n \geq k$. Let $M := max \{|x_1|, |x_2|, \dots, |x_k-1|, 1 + |x|\}$, Hence, $|x_n| \leq M$ for all $n \in N$.

## Operations of Sequences

Let $X = \langle x_n \rangle$ and $Y = \langle y_n \rangle$ be sequences. Then

- **SUM**: $X + Y := \langle x_n + y_n \rangle$
- **DIFFERENCE**: $X - Y := \langle x_n - y_n \rangle$

- **QUOTIENT**: $\frac{X}{Y} := \left\langle \frac{x_n}{y_n} \right\rangle$, where $y_n \neq 0 \forall n$

$$
\begin{align}

# Introduction to Infinite Series

If $X := \langle x_n \rangle$ is a sequence in $\mathbb{R}$, then the ***infinite series*** (or sinply ***series***) **generated by $X$** is the sequence $S := \langle S_k \rangle$ defined by

$$
\begin{align}
S_1 & := x_q \\
S_2 & := S_q + x_1 && (= x_1 + x_2) \\
\vdots & \\
S_k & := S_{k - 1} + x_k && (= x_1 + x_2 + \dots + x_k) \\
\vdots & \\
\end{align}
$$

The numbers $x_n$ are called the ***terms*** of the series and the numbers $S_k$ are called the ***partial sums*** of this series (is ***sums*** of this series). If $\lim S$ exists, we say that this series is ***convergent*** and call this limit the **sum** or the **value** of this series. If this limit does not exist, we say the series is ***divergent***.

The series is denoted by

$$
\sum \langle x_n \rangle \text{or} \sum x_n \text{or} \sum_{n = 1}^{\infty} x_n
$$

**Example 1**

Consider the sequence $X := \langle r^n \rangle_{n = 0}^{\infty}$ where $r \in \mathbb{R}$, which generates the **geometric series**:

$$
\sum_{n = 0}^{\infty} r^n = 1 + r + r^2 + \dots + r^n + \dots
$$

If $|r| < 1$, then the series converges to $\frac{1}{1 - r}$.

**Note**

If **geometric series** is having $|r| < 1$, then it is enough to use a partial sum to get the value of the infinite series, that is,

$$
S_n = \frac{a_1}{1 - r}
$$

where $a_1$ is the first term of a sequence and $r$ is the common ratio.

**Example 2**

Find the sum of this serires $\sum \frac{1}{2^n}$. Note that $r = \frac{1}{2}$. Then $|r| = \frac{1}{2} < 1$. Thus,

$$
\begin{align}
\sum \frac{1}{2^n} & = S_n \\
& = \frac{\frac{1}{2}}{1 - \frac{1}{2}} \\
& = 1
\end{align}
$$

**Example 3**

Consider the series generated by $\langle (-1)^n \rangle$, that is the series

$$
\sum_{n = 0}^{\infty} (-1)^n = (+1) + (-1) + (+1) + (-1) + \dots
$$

Then the sequence of partial sums is $(1, 0, 1, 0, 1, 0, \dots)$. Thus, the series $\sum (-1)^n$ is divergent.

## Theorem 1 *The $n$th Term Test*

If the series $\sum x_n$ converges, the $\lim x_n = 0$.

**Proof**

By the definition, the convergence of $sum x_n$ requires that $\lim S_k$ exists. Since $x_n = S_n - S_{n - 1}$, it follows that $\lim x_n = \lim S_n - \lim S_{n - 1} = 0$.

**Example 1**

Consider the sequence $X := \langle r^n \rangle_{n = 0}^{\infty}$ where $r \in \mathbb{R}$, which generates the **geometrix series**:

$$
\sum_{n = 0}^{\infty} r^n = 1 + r + r^2 + \dots + r^n + \dots
$$

If $|r| \geq 1$, then the geometric series diverges.

**Example 2**

- The **harmonic series** $\sum \frac{1}{n}$ diverges.
- The **$2$-series** $\sum \frac{1}{n^2}$ is convergent.
- The $p$**-series** $\sum \frac{1}{n^p}$ converges when $p > 1$.
- The $p$**-series** $\sum \frac{1}{n^p}$ diverges when $0 < p \leq 1$.

# Infinite Series

For the infinite series $\sum x_n$, the $n$***th partial sum*** is given by

$$
S_n = x_1 + x_2 + x_3 + \dots + x_n
$$

If the sequence of the partial sum $\langle S_n \rangle$ converges to $S$, then the series $\sum x_n$ converges. The limit $S$ is called the ***sum of the series***, i.e.,

$$
S = x_1 + x_2 + \dots + x_n + \dots
$$

If $\langle S_n \rangle$ diverges, then the series diverges.

## Geometric Series

If **geometric series** $\sum x_n$ is having $|r| < 1$, then the series is convergest to $S_n$, that is,

$$
\sum x_n \to S_n = \frac{a_1}{1 - r}
$$

where $a_1$ is the first term of a sequence and $r$ is the common ration. Otherwise, the series is divergent.

## Properties of Infinite Series

If $\sum x_n = X$, $\sum y_n = Y$ and $c \in \mathbb{R}$, then the following series converge to the indicated sums.

- $\sum cx_n = cX$
- $\sum (x_n + y_n) = X + Y$
- $\sum (x_n - y_n) = X - Y$

## The $n$th Term Test

If $\sum x_n converges, then $\lim x_n = 0$.

## Theorem 1 ***The $n$th Term Test for Divergence***

If $\lim x_n \neq 0$, then $\sum x_n$ diverges.

## Theorem 2 **Direct Comparison Test**

Let $0 < x_n \leq y_n$ for all $n$.

- If $\sum y_n$ converges, then $\sum x_n$ converges.
- If $\sum x_n$ diverges, then $\sum y_n$ diverges.

## Theorem 3 **Alternating Series Test**

Let $x_n > 0$ The alternating series

$$
\sum_{n = 1}^{\infty} (-1)^n x_n \text{ and } \sum_{n = 1}^{\infty} (-1)^n x_n
$$

converges if the following conidtion are true.

- $\lim a_n = 0$
- $x_n + 1 \leq x_n$ for all $n$

**Example 1**

Show that $\sum \frac{(-1)^n}{n}$is a convergest.

**Proof**

$$
\frac{(-1)^n}{n} = \sum (-1)^n \left( \frac{1}{n} \right)
$$

Since $\lim \frac{1}{n} = 0$ and $x_n + 1 \leq x_n$ for all $n$, by Theorem 3 $\sum \frac{(-1)^n}{n} is convergest.

**Example 2**

Determine whether $\sum \frac{n^2 + 2}{n^4 + 5} converges or diverges.

**Solution**

$$
\begin{align}
\sum \frac{n^2 + 2}{n^4 + 5} & < \sum \frac{n^2 + 2}{n^4} \\
& = \sum \frac{n^2}{n^4} + \sum \frac{2}{n^4} \\
& = \sum \frac{1}{n^2} + 2 \sum \frac{1}{n^4}
\end{align}
$$

Note that $\sum \frac{1}{n^2}$ and $\sum \frac{1}{n^4}$ are both $p$-series and both converges. Thus, $\sum \frac{1}{n^2} + 2 \sum \frac{1}{n^4}$ converges. Thus, by Comparison Test $\sum \frac{n^2 + 2}{n^4 + 5}$ converges.

## Theorem 4

If the series $\sum |a_n|$ converges, then the series $\sum a_n$ is converges.

**Proof**

Let $\langle a_n \rangle$ be a sequence. Consider $\sum |a_n|$ convergence. Thus $\sum |a_n| = \lim |S_k|$, where $|S_k| = \sum_{n = 1}^k |a_n|$. Thus, $\left| |S_k| - L| \right| < \epsilon (\epsilon > 0, L \in \mathbb{R})$. Let $S_k = \sum_{n=1}^k a_n$. Now,

$$
\begin{align}
|S_k - L| & = |(a_1 + a_2 + \dots + a_k) - L| \\
& \leq ||(a_1 + a_2 + \dots + a_k| - L| \\
& \leq |(|a_1| + |a_2| + \dots + |a_k|) - L| \\
& = ||S_k| - L| \\
& < \epsilon
\end{align}
$$

Thus, $\lim S_k$ exists. Hence, $\sum a_n$ converges.

## Theorem 5 *Absolute and Conditional Convergence

- $\sum a_n$ is ***absolutely convergent*** if $\sum |a_n|$ converges.
- $\sum a_n$ is ***conditional convergent*** if $\sum a_n$ converges but $\sum |a_n|$ divergest.

**Example**

Let $\langle x_n \rangle = \left\langle \frac{(-1)^n}{n} \right\rangle$. Then by alternating series test, $\sum \frac{(-1)^n}{n}$ converges. But,

$$
\begin{align}
\sum \left| \frac{(-1)^n}{n} \right| & = \sum \frac{|(-1)^n|}{|n|} \\
& = \sum \frac{|-1|^n}{n} \\
& = \sum \frac{1}{n} \\
& = \infty
\end{align}
$$

Thus, $\sum \frac{(-1)^n}{n}$ is a conditional convergent.

## Theorem 6 *Ratio Test*

Let $\sum a_n$ be the series with nonzero terms.

- $\sum a_n$ converges absolutly if $\lim \left| \frac{a_n + 1}{a_n} \right| < 1$.
- $\sum a_n$ diverges if $\lim \left| \frac{a_n + 1}{a_n} \right| > 1$.
- The Ratio Test is inconclusive if $\lim \left| \frac{a_n + 1}{a_n} \right| = 1$.

**Example**

Determine whether $\sum_{n = 2}^{\infty} \frac{n^2}{(2n - 1)!}$ converges or diverges.

**Solution**

$$
\begin{align}
\lim \left| \frac{((n + 1)^2)}{(2(n + 1) - 1)!} \cdot \frac{(2n - 1)!}{n^2} \right| & = \lim \left| \frac{((n + 1)^2)}{(2n + 1)!} \cdot \frac{(2n - 1)!}{n^2} \right| \\
& = \lim \frac{n^2 + 2n + 1}{(2n + 1)(2n)(2n -1)!} \cdot \frac{(2n - 1)!}{n^2} \\
& = \lim \frac{n^2 + 2n + 1}{4n^4 + n^2} = 0 < 1
\end{align}
$$

By Ratio Test, $\sum_{n = 2}^{\infty} \frac{n^2}{(2n - 1)!}$ converges.

## Theorem 7 *Root Test*

Let $\sum a_n$ be the series.

- $\sum a_n$ converges absolutly if $\lim \sqrt[n]{|a_n|} < 1$.
- $\sum a_n$ diverges if $\lim \sqrt[n]{|a_n|} > 1$ or $\lim \sqrt[n]{|a_n|} = \infty$.
- The Root Test is inconclusive if $\lim \sqrt[n]{|a_n|} = 1$.

**Note**

$$
\lim n^{\frac{1}{n}} = 1
$$

**Example 1**

Determine whether $\sum \frac{n^n}{3^{1 + 2n}}$ converges or diverges.

**Solution**

$$
\begin{align}
\lim \left| \frac{n^n}{3^{1 + 2n}} \right|^\frac{1}{n} & = \lim \frac{n}{3^{\frac{1}{n} + 2}} \\
& = \infty
\end{align}
$$

Thus, by Root Test $\sum \frac{n^n}{3^{1 + 2n}}$ diverges.

**Example 2**

Determine whether $\sum_{n = 3}^{\infty} \frac{(-12)^n}{n}$ converges or diverges.

**Solution**

$$
\begin{align}
\lim \left| \frac{(-12)^n}{n} \right|^{\frac{1}{n}} & = \lim \frac{12}{n^{\frac{1}{n}}} \\
& = \frac{12}{1} = 12 \\
& > 1
\end{align}
$$

Hence, by Root Test $\sum_{n = 3}^{\infty} \frac{(-12)^n}{n}$ diverges.

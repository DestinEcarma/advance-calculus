# Derivatives

The rate of change, or slope, of a function called its ***derivative***. It is denoted by $f'(x)$, which is read as $f$ prime of $x$.

$$
f'(x) = \lim_{h \to 0} \frac{f(x + h) - f(x)}{h}
$$

1. The derivative is an equation for the slope of the tangent line at any point $(x, f(x))$
2. If $f'(x) exists for some value $x$, then we say $f$ is *differentiable* at $x$.
3. A function differentiable at every point in its domain is a ***differentiable function***.

The derivative of $f$ at $x = a$ can also be found by

$$
f'(a) = \lim_{x \to a} \frac{f(x) - f(a)}{x - a}
$$

**Example 1**

Find the derivative of $f(x) = 4x^2$.

**Solution**

$$
\begin{align}
f'(x) & = \lim_{h \to 0} \frac{f(x + h) - f(x)}{h} \\
& = \lim_{h \to 0} \frac{4(x + h)^2 - 4x^2}{h} \\
& = \lim_{h \to 0} \frac{4(x^2 + 2xh + h^2) - 4x^2}{h} \\
& = \lim_{h \to 0} \frac{4x^2 + 8xh + 4h^2 - 4x^2}{h} \\
& = \lim_{h \to 0} \frac{8xh + 4h^2}{h} \\
& = \lim_{h \to 0} \frac{h(8x + 4h)}{h} = \lim_{h \to 0} 8x + 4h = 8x + 0 = 8x
\end{align}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 500)
y1 = 4 * x**2
y2 = 8 * x

plt.plot(x, y1, label=r"$y=4x^2$", color="red")
plt.plot(x, y2, label=r'$y = 8x$', color='blue')

plt.xlabel("x")
plt.ylabel("y")
plt.title(r"Graph Illustration")

plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)

plt.ylim(-20, 20)
plt.xlim(-5, 5)

plt.grid(True, linestyle="--", alpha=0.6)

plt.legend()
plt.show()
```

**Example 2**

Find the derivative of $y = x^3$.

**Solution**

$$
\begin{align}
y' & = \lim_{h \to 0} \frac{(x + h)^3 - x^3}{h} \\
& = \lim_{h \to 0} \frac{x^3 + 3x^2h + 3xh^2 + h^3 - x^3}{h} \\
& = \lim_{h \to 0} \frac{3x^2h + 3xh^2 + h^3}{h} \\
& = \lim_{h \to 0} \frac{h(3x^2 + 3xh + h^2)}{h} \\
& = \lim_{h \to 0} (3x^2 + 3xh + h^2) \\
& = 3x^2 + 3x(0) + (0)^2 = 3x^2
\end{align}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 500)

y1 = x ** 3
y2 = 3 * x ** 2

plt.plot(x, y1, label=r"$x^3$", color="red")
plt.plot(x, y2, label=r"$3x^2$", color="blue")

plt.xlabel("x")
plt.ylabel("y")
plt.title(r"Graph Illustration")

plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)

plt.ylim(-40, 40)
plt.xlim(-5, 5)

plt.grid(True, linestyle="--", alpha=0.6)

plt.legend()
plt.show()
```

## Theorem 1 *Power Rule*

$$
\begin{align}
\frac{d}{dx} (x^n) & = nx^{n - 1} & (n \text{ is the exponent or power of } x)
\end{align}
$$

**Example 1**

Find the derivative of $y = \frac{1}{4} x^4$.

**Solution**

$$
\begin{align}
y & = \frac{1}{4} x^4 \\
\frac{dy}{dx} & = \left( \frac{1}{4} \right) (4x^{4 - 1}) \\
& = x^3
\end{align}
$$

**Example 2** ***Approximation***

Approximate the value of $\sqrt{0.037}$ using derivatives.

**Solution**

Let consider a function $f(x) = \sqrt{x}$. Then

$$
f'(x) = \frac{1}{2} x^{-\frac{1}{2}}
$$

Note that $0.037 = 0.04 - 0.003$, $h = -0.003$ and $x = 0.04$. Now,

$$
f'(x) = \frac{f(x + h) - f(x)}{x}
$$

This implies that $f(x + h) = f(x) + hf'(x)$. Then

$$
\begin{align}
f(0.04 + (-0.003)) & = f(0.04) + (-0.003) \left( \frac{1}{2} (0.04)^{-\frac{1}{2}} \right) \\
& = \sqrt{0.04} - \left( \frac{3}{1000} \right) \left( \frac{1}{2} \right) \left( \frac{1}{\sqrt{0.04}} \right)
\end{align}
$$

Or

$$
\begin{align}
f(0.004 + (-0.003)) & = 0.2 - \left( \frac{3}{1000} \right) \left( \frac{1}{2} \right) \left( \frac{1}{0.2} \right) \\
& = \frac{20}{100} - \left( \frac{3}{1000} \right) \left( \frac{1}{2} \right) \left( \frac{1}{\frac{2}{10}} \right) \\
& = \frac{20}{100} - \frac{3}{400} \\
& = \frac{77}{400}
\end{align}
$$

Thus, $f(0.037) \approx 0.1925$.

Hence, $\sqrt{0.037} \approx 0.1925$.

**Activity 1**

Find the local maxima and local minima of the function

$$
2x^3 - 21x^2 + 36x - 20
$$

**Solution**

Find the first derivative

$$
\begin{align}
\frac{d}{dx}(2x^3) & = 3(2x^{3 - 1}) = 6x^2 \\
\frac{d}{dx}(-21x^2) & = 2(-21x^{2 - 1}) = -42x \\
\frac{d}{dx}(36x) & = 1(36x^{1 - 1}) = 36 \\
\frac{d}{dx}(-20) & = 0 & (\text{The derivative of a constant is always zero})\\
\end{align}
$$

Thus,

$$
f'(x) = 6x^2 - 42x + 36
$$

We then solve for critical points by setting $f'(x) = 0$

$$
6x^2 - 42x + 36 = 0
$$

This is a quadratic equation.

$$
\begin{align}
\implies & 6x^2 - 42x + 36 = 0 \\
\implies & x^2 - 7x + 6 = 0 \\
\implies & (x - 6)(x - 1) \\
\implies & x = 6, x = 1
\end{align}
$$

Next we determine the maxima and minima using the second derivative. To find the second derivative we simply differentiate the first derivative,

$$
\begin{align}
\frac{d}{dx}(6x^2) & = 2(6x^{2 - 1}) = 12x \\
\frac{d}{dx}(-42x^2) & = 1(-42x^{1 - 1}) = -42
\end{align}
$$

Thus,

$$
f''(x) = 12x - 42
$$

This will help us determine the concavity.

Evaluate for $x = 1$

$$
\begin{align}
f''(1) & = 12(1) - 42 = 12 - 42 = -30 & (\text{Concave down, local maxima})
\end{align}
$$

Evaluate for $x = 6$

$$
\begin{align}
f''(6) & = 12(6) - 42 = 72 - 42 = 3 & (\text{Concave up, local minima})
\end{align}
$$

Lastly, find the corresponding values from the given function,

$$
\begin{align}
f(1) & = 2(1^3) - 21(1^2) + 36(1) - 20 = 2 -21 + 36 - 20 = -3 \\
f(6) & = 2(6^3) - 21(6^2) + 36(6) - 20 = 432 - 756 + 216 - 20 = 128
\end{align}
$$

Therefore, the local maximum for $x = 1$ with $f(1) = -3$, and the local minimum for $x = 6$ with $f(6) = -128$.

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 2 * x ** 3 - 21 * x ** 2 + 36 * x - 20

x = np.linspace(-10, 10, 500)

y1 = f(x)
y2 = f(1)
y3 = f(6)
y4 = 6 * x ** 2 - 42 * x + 36

plt.plot(x, y1, label=r"$2x^3 - 21x^2 + 36x - 20$", color="red")
plt.plot(x, y4, label=r"$6x^2 - 42x + 36$", color="gray")

plt.plot(1, y2, label=f"Maxima: ${y2}$", color="blue", marker="o", markersize=8)
plt.plot(6, y3, label=f"Minima: ${y3}$", color="green", marker="o", markersize=8)

plt.xlabel("x")
plt.ylabel("y")
plt.title(r"Graph Illustration")

plt.axhline(0, color="black", linewidth=1)
plt.axvline(0, color="black", linewidth=1)

plt.ylim(-150, 150)
plt.xlim(-15, 15)

plt.grid(True, linestyle="--", alpha=0.6)

plt.legend()
plt.show()
```

Find the derivative of $f(x) = 4$

**Solution**

The simplest answer is to say, because we have a constant derivative hence, the value is **zero**. However, to understand **why** is it a zero, we can apply the definition of the derivative using the following equation using limits,

$$
\begin{align}
\lim_{h \to 0} \frac{f(x + h) - f(x)}{h} & \to \frac{4 - 4}{h} \\
& \to \frac{0}{h} \\
& \to 0
\end{align}
$$
